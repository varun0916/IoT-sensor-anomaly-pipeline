```python
# List of your Bronze tables as per the screenshot
bronze_tables = [
    "sensor_readings", 
    "operations_table", 
    "environment_table", 
    "time_events", 
    "health_diagnostics"
]

def test_bronze_layer():
    print("🚀 Starting Bronze Layer Validation...")
    for table in bronze_tables:
        full_table_name = f"iot_catalog_p2.bronze.{table}"
        try:
            count = spark.table(full_table_name).count()
            assert count > 0, f"Table {table} is empty!"
            print(f"✅ Pass: {table} contains {count} records.")
        except Exception as e:
            print(f"❌ Fail: Error accessing {table}: {str(e)}")

test_bronze_layer()
```
